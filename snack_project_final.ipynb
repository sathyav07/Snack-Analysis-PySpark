Read excel file

```python
import pandas as pd

mcd = pd.read_excel("menu.xlsx")
starbucks_food = pd.read_excel("starbucks-menu-nutrition-food.xlsx")
starbucks_drinks = pd.read_excel("starbucks-menu-nutrition-drinks.xlsx")

# Rename problematic columns in starbucks_food
starbucks_food = starbucks_food.rename(columns={
    'Fat (g)': 'Fat_g',
    'Carb. (g)': 'Carb_g',
    'Fiber (g)': 'Fiber_g',
    'Protein (g)': 'Protein_g'
})

# Rename problematic columns in starbucks_drinks
starbucks_drinks = starbucks_drinks.rename(columns={
    'Fat (g)': 'Fat_g',
    'Carb. (g)': 'Carb_g',
    'Fiber (g)': 'Fiber_g',
    'Protein': 'Protein_g' # Assuming 'Protein' in drinks should also be consistent
})
```

Create spark session

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Snack Analysis") \
    .getOrCreate()
```

Convert Pandas DataFrame to PySpark DataFrame

```python
mcd_spark = spark.createDataFrame(mcd)

starbucks_food_spark = spark.createDataFrame(starbucks_food)

starbucks_drinks_spark = spark.createDataFrame(starbucks_drinks)
```

Verify the data

```python
mcd_spark.show(5)
```

Check column names

```python
print(mcd_spark.columns)

print(starbucks_food_spark.columns)

print(starbucks_drinks_spark.columns)
```

# Average Calories

```python
from pyspark.sql.functions import avg

print("McDonald's Average Calories")
mcd_spark.select(avg("Calories").alias("Avg Calories")).show()

print("Starbucks Average Calories")
starbucks_food_spark.select(avg("Calories").alias("Avg Calories")).show()
```

# Average Fat

```python
print("McDonald's Average Fat")
mcd_spark.select(avg("Total Fat").alias("Avg Fat")).show()

print("Starbucks Average Fat")
starbucks_food_spark.select(avg("Fat_g").alias("Avg Fat")).show()
```

# Average Protein

```python
print("McDonald's Average Protein")
mcd_spark.select(avg("Protein").alias("Avg Protein")).show()

print("Starbucks Average Protein")
starbucks_food_spark.select(avg("Protein_g").alias("Avg Protein")).show()
```

```python
from pyspark.sql.functions import avg, when, col

print("Starbucks Drinks Average Sodium")
starbucks_drinks_spark.select(
    avg(when(col("Sodium") == "-", None).otherwise(col("Sodium")).cast("double")).alias("Avg Sodium")
).show()
```
